# Simulação do pêndulo elástico

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from matplotlib.animation import FuncAnimation
from IPython import display
```

### Parâmetros:

```python
# comprimento de equilíbrio da mola
r0 = 1
# gravidade
g = 9.85
# constante da mola
k = 4*g
# massa
m = 1
```

### Derivadas:

```python
def deriv(y, t, r0, k, m):
    """Return the first derivatives of y = theta, z1, L, z2."""
    theta, z1, r, z2 = y

    thetadot = z1
    z1dot = (-g*np.sin(theta) - 2*z1*z2) / r
    rdot = z2
    z2dot = (m*r*z1**2 - k*(r-r0) + m*g*np.cos(theta)) / m
    return thetadot, z1dot, rdot, z2dot
```

```python
def solu(y0, tmax = 10, dt = 0.01):
    "Resolve numéricamente as equações difernciais a partir dos valores iniciais de theta, thetadot, r, rdot"
    t = np.arange(0, tmax+dt, dt)
    #integrando as equações no intervalo de tempo:
    y = odeint(deriv, y0, t, args=(r0, k, m))
    theta, r = y[:,0], y[:,2]
    #convertendo para coordenadas cartesianas:
    x = r * np.sin(theta)
    y = -r * np.cos(theta)
    #gráfico da solução:
    fig, ax = plt.subplots(1,1)
    ax.set_title('Solução Numérica')
    ax.plot(x,y, label = 'Tragetória Calculada')
    ax.plot(0,0,'o', c = 'k', label = 'Eixo do Pêndulo')
    ax.plot(x[0], y[0], 'o', c = 'r', label = f'Posição Inicial: x = {x[0]:.3f} e y = {y[0]:.3f}')
    ax.legend(fontsize='small')
    ax.set_xlim(-r0,r0)
    ax.set_xlabel('x[m]')
    ax.set_ylabel('y[m]')
    plt.savefig('imagens/exemplo_ressonancia.png',dpi = 200)
    plt.show()
    #retorna os valores da solução
    return x,y, t, dt
```
